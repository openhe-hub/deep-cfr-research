```python
import matplotlib.pyplot as plt
import numpy as np
import os

folder = '../assets/adv_loss/'
```

```python
def calc_avg_loss(files, limit=10):
    avg_arr = []
    for file in files:
        losses = []
        avg_losses = []
        with open(os.path.join(folder, file), 'r') as fp:
            iter_data = []
            for line in fp:
                line = line.strip()
                if line.startswith('==='):
                    if iter_data:
                        losses.append(iter_data)
                        # print(iter_data)
                    iter_data = []
                else:
                    iter_data.append(float(line))
            
            if iter_data:
                losses.append(iter_data)
                # print(iter_data)

        for arr in losses:
            avg_losses.append(np.mean(np.array(arr)))

        print(file)
        print(len(losses), losses)
        print(len(avg_losses), avg_losses)

        avg_arr.append(avg_losses[:limit])
    return avg_arr
```

```python
plt.figure(figsize=(10, 6))

avg_losses = calc_avg_loss(['adv_loss_0.txt'], limit=20)
labels = ['ResV2']
for losses, label in zip(avg_losses, labels):
    plt.plot(losses, label=label, marker='s')

plt.title(f'DEEP-CFR ADV MODEL(FC)')
plt.xlabel('Epochs')
plt.ylabel('MSE Loss')
plt.legend()

plt.grid(True)
# plt.savefig(f'../assets/adv_loss/avg_adv_loss_{player_id}.jpg')
plt.show()
```
