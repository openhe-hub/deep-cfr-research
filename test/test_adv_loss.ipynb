```python
import matplotlib.pyplot as plt
import os

player_id = 1
folder = '../assets/adv_loss/'
```

```python
losses = []
with open(os.path.join(folder, f'adv_loss_{player_id}.txt'), 'r') as fp:
    iter_data = []
    for line in fp:
        line = line.strip()
        if line.startswith('==='):
            if iter_data:
                losses.append(iter_data)
            iter_data = []
        else:
            iter_data.append(float(line))
    
    if iter_data:
        losses.append(iter_data)

print(losses)
```

```python
selected_iters = [0, 30, 60, 90, 120, 150]

plt.figure(figsize=(10, 6))

for iter_index in selected_iters:
    plt.plot(losses[iter_index], label=f'ITER {iter_index}', marker='s')

plt.title(f'DEEP-CFR ADV MODEL(player {player_id})')
plt.xlabel('Epochs(per 200)')
plt.ylabel('MSE Loss')
plt.legend()

plt.grid(True)
plt.savefig(f'../assets/adv_loss/adv_loss_{player_id}.jpg')
plt.show()
```
