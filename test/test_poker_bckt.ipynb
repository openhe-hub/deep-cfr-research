```python
output_record = []
NUM_2_INDEX = {
    2: 0,
    5: 1,
    6: 2,
    7: 3,
}
output_record_path = f'../assets/slumbot/output_record.txt'
with open(output_record_path, "r") as file:
    data = file.readlines()
output_record = [ast.literal_eval(line.strip()) for line in data]
len(output_record)
```

```python
import re

def transform(cards, idx):
	num_dict = {
		"s": 0,
		"h": 0,
		"c": 0,
		"d": 0,
	}
	
	for card in cards:
		num_dict[card[-1]] += 1
	
	max_suit = max(num_dict, key=num_dict.get)
	max_suit_count = num_dict[max_suit]

	MIN_NUM = {2: 2, 5: 3, 6: 4, 7: 5}
	if max_suit_count < MIN_NUM[idx]:
		return ''.join([f"{card[:-1]}" for card in cards])
	else:
		result = []
		for card in cards:
			suit = card[-1]
			if suit == max_suit:
				result.append(f"{card[:-1]}s") 
			else:
				result.append(f"{card[:-1]}o") 
		return ''.join(result)

def find(code, idx):
	def get_card_rank(card):
		ranks = '23456789TJQKA'
		return ranks.index(card)
	
	perm = ""
	if len(code) == idx:
		hand = ''.join(sorted(code[:2], key=get_card_rank))
		remaining = ''.join(sorted(code[2:], key=get_card_rank))
		perm = hand + remaining

	elif len(code) == idx*2:
		grouped = [code[i:i+2] for i in range(0, len(code), 2)]
		hand = ''.join(sorted(grouped[:2], key=lambda x: (get_card_rank(x[0]), x[1] )))
		remaining = ''.join(sorted(grouped[2:], key=lambda x: (get_card_rank(x[0]), x[1]) ))
		perm = hand + remaining

	else:
		return None
	
	if perm in ref_data[idx]:
		return [perm, ref_data[idx][perm]]

	print(f"{code}, {perm} not found")
	
	return None

def get_curr_street_action(path, IDX, is_first, hand_win):
	splits = path.split('/')
	curr_splits = splits[-1]
	if curr_splits.endswith('f') and hand_win < 0:
		return 'f'
	elif curr_splits == 'f' and hand_win > 0:
		return '/'
	result = re.split(r'[kbfc]+', curr_splits) 
	matches = list(re.finditer(r'[kbfc]', curr_splits))	
	if len(matches) <= 0: 
		return ''
	action1 = matches[1].group()
	action2 = matches[0].group()
	return action1 if is_first else action2

IDX = 7
side = 'model' # or slumbot
cnts = [0 for i in range(8)]
my_data = [[] for i in range(8)]
for i, hand_data in enumerate(total_data):
	cards = []
	hand_win = 0
	if side == 'model':
		cards = hand_data['hole_cards'] + hand_data['board_cards']
		hand_win = hand_data['hand_win']
	elif side == 'slumbot':
		cards = hand_data['bot_hole_cards'] + hand_data['board_cards']
		hand_win = -hand_data['hand_win']
	# stats
	idx = len(cards)
	cnts[idx] += 1
	if idx >= IDX:
		cards = cards[:IDX]
		code = transform(cards, IDX)
		res = find(code, IDX)
		if res == None:
			continue
		group = res[-1]

		action = output_record[i][NUM_2_INDEX[IDX]][0] if len(output_record[i][NUM_2_INDEX[IDX]]) else '/'
		# action = get_curr_street_action(hand_data['history_action'], IDX, hand_data['is_first'], hand_win)
		if action == '/': continue
		
		my_data[IDX].append({
			'my_hole_cards': cards[:2],
			'slumbot_hole_cards': hand_data['bot_hole_cards'],
			'board_cards': hand_data['board_cards'],
			'hand_win': hand_win,
			'code': code,
			'group': group,
			'is_first': hand_data['is_first'],
			'action_path': hand_data['history_action'],
			'target_action_path': action,
		})
		
len(my_data[IDX])
```

```python
import pandas as pd
import os
from pathlib import Path

def combine_csv_to_excel(folder_path, output_excel='combined_data.xlsx'):
    if not os.path.exists(folder_path):
        raise ValueError(f"{folder_path} not exist")
    
    csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
    
    if not csv_files:
        print("no csv!")
        return
    
    with pd.ExcelWriter(output_excel) as writer:
        for csv_file in csv_files:
            try:
                file_path = os.path.join(folder_path, csv_file)
                df = pd.read_csv(file_path)
                
                sheet_name = Path(csv_file).stem
                
                if len(sheet_name) > 31:
                    sheet_name = sheet_name[:31]
                
                df.to_excel(writer, sheet_name=sheet_name, index=False)
                print(f"handled: {csv_file}")
                
            except Exception as e:
                print(f"error while handling {csv_file}: {str(e)}")
    
folder_path = "../assets/slumbot/stat/stat-8"
output_path = "../assets/slumbot/stat/stat-8/stat-res-v3-b3.xlsx"
combine_csv_to_excel(folder_path, output_path)
```
